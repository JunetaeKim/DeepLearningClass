```python
import numpy as np
import matplotlib.pyplot as plt
```

### Linear function

```python
# a linear function 
x = np.arange(-4, 5)
y = 2*x + 1

# Plotting the function 
plt.figure(figsize=(7,5))
plt.plot(x, y)
plt.scatter(x, y, c='red')
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
```

### Quadratic function1

```python
# a quadratic function 
x = np.arange(-5, 5.2, 0.2)
y = 2*x**2 + 1

# Plotting the function 
plt.figure(figsize=(7,5))
plt.plot(x, y)
plt.scatter(x, y, c='red')
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
```

### Quadratic function2

```python
# a quadratic function 
x = np.arange(-5, 5.2, 0.2)
y = -2*x**2 + 1

# Plotting the function 
plt.figure(figsize=(7,5))
plt.plot(x, y)
plt.scatter(x, y, c='red')
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
```

### Natural exponential function

```python
# an exponential  function 
x = np.arange(-5, 5.2, 0.2)
y = np.exp(x)

# Plotting the function 
plt.figure(figsize=(7,5))
plt.plot(x, y)
plt.scatter(x, y, c='red')
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
```

### Log function with domain from 0.01 to 1

```python
# a log function 
x = np.arange(0.01, 1., 0.05)
y = np.log(x)

# Plotting the function 
plt.figure(figsize=(7,5))
plt.plot(x, y)
plt.scatter(x, y, c='red')
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
```

### Log function shifted horizontally by 1 on the x-axis

```python
# a log function 
x = np.arange(0.01, 1., 0.05)
y = np.log(1-x)

# Plotting the function 
plt.figure(figsize=(7,5))
plt.plot(x, y)
plt.scatter(x, y, c='red')
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
```

### Sigmoid function

```python
# a sigmoid function 
x = np.arange(-5., 5., 0.2)
y = 1/(1+np.exp(-x))

# Plotting the function 
plt.figure(figsize=(7,5))
plt.plot(x, y)
plt.scatter(x, y, c='red')
plt.axhline(0, color='black', linestyle='--')
plt.axhline(1, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
```
