```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def derivative (x, n, h):
    return ((x+h)**n - x**n)/h

derivativeRes = []
hValues = [10, 1, 0.1, 0.01, 0.001, 0.0001, 0.00001, 0.000001]
for i in hValues:
    derivativeRes.append(derivative(10, 2, i))
    

plt.plot(hValues, derivativeRes)
plt.ylabel('Derivative')
plt.xlabel('h value')
```

```python
def Natural_e (n):
    return (1+(1/n))**n

Natural_e_Res = []
n = [0.001, 0.01, 0.1, 1, 10, 100, 1000, 10000]

for i in n:
    Natural_e_Res.append(Natural_e(i))

plt.plot(n, Natural_e_Res)
plt.ylabel('e')
plt.xlabel('n value')
```

```python
def derivative (h):
    return ((2.71828)**h-1)/h

derivativeRes = []
hValues = [2, 1, 0.1, 0.01, 0.001, 0.0001, 0.00001, 0.000001]

for i in hValues:
    derivativeRes.append(derivative(i))

plt.plot(derivativeRes)
plt.ylabel('Derivative')
plt.xlabel('h value')
```

```python
def derivative (x,  h):
    return (2.71828**(x+h) - 2.71828**x)/h

derivativeRes = []
ExpRes = []

xValues = [1,2,3,4]

for i in xValues:
    derivativeRes.append(derivative(i, 0.000001))
    
for i in xValues:
    ExpRes.append(2.71828**i)
    

plt.plot(derivativeRes)
plt.plot(ExpRes)
plt.ylabel('Derivative')
plt.xlabel('x value')
```

```python
def derivative (x,  h):
    return (np.log(x+h) - np.log(x))/h

derivativeRes = []
xValues = [1,2,3,4]
for i in xValues:
    derivativeRes.append(derivative(i, 0.000001))

LnRes = []
for i in xValues:
    LnRes.append(1/i)
    
plt.plot(xValues, derivativeRes)
plt.plot(xValues, LnRes)

plt.ylabel('Derivative')
plt.xlabel('x value')
```
