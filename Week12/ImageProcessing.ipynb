```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import sys
import requests
```

### Read image data

```python
url = 'https://raw.githubusercontent.com/JunetaeKim/DeepLearningClass/main/Week12/choonsik.jpg'
Fig = Image.open(requests.get(url, stream=True).raw)
Fig
```

### Resize image data

```python
Fig = Fig.resize((220,220))
Fig
```

### Convert Image to Grayscale

```python
GFig = Fig.convert('L')
GFig
```

### Convert Image to numpy

```python
FigNP = np.array(Fig)
FigNP.shape
```

### Showing figure from numpy file

```python
plt.figure(figsize=(7,7))
plt.subplot(2,2,1)
plt.imshow(FigNP, cmap='Greys')
    
for i in range(0,3):
    split_img = np.zeros(FigNP.shape, dtype="uint8")

    print(i)
    split_img[:,:,i]=FigNP[:,:,i]
    
    plt.subplot(2,2,i+2)
    plt.imshow(split_img)
```

### Convert Image to numpy and resize it again

```python
url = 'https://raw.githubusercontent.com/JunetaeKim/DeepLearningClass/main/Week12/five.png'
Five = Image.open(requests.get(url, stream=True).raw)
plt.imshow(Five)

Five = Five.resize((15,35))
FigNP = np.array(Five)
FigNP.shape
```

### Print numpy

```python
for i in np.array(FigNP)[:, :, 1]:
    for j in i:
        sys.stdout.write('%d\t' % j)
    sys.stdout.write('\n')
```
