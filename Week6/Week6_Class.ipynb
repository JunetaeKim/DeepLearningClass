```python
import numpy as np
import matplotlib.pyplot as plt
```

## Create a class that calculates factorial

```python
class Factorial:
    
    def __init__(self, Num):
        self.Num = Num
        
    def DoFactorial(self,):
        self.FactRes = self.Num
        
        for i in range(1, self.Num):
            self.FactRes *= (self.Num-i)
        
```

### Instantiation

```python
Fac = Factorial(5)
```

```python
Fac.Num
```

### Do factorial

```python
Fac.DoFactorial()
Fac.FactRes
```

## Find the minimum value of 0

#### Without class

```python
x = np.arange(-10, 10, 0.2)
y = x**2
plt.plot(x, y  )

for Ind in range(40):
    plt.scatter(x[Ind], y[Ind])
```

```python
class MinZero:
    
    def __init__(self, ):
        self.Iter = 0
        
        self.x = np.arange(-10, 10, 0.2)
        self.y = self.x**2
        
    def FindMin(self,NumIter):
        self.NumIter = NumIter
        
        if NumIter <0:
            Dir = -1
        else:
            Dir = 1
            

        plt.plot(self.x, self.y)
        for Ind in range(self.Iter, self.Iter + self.NumIter, Dir):
            plt.scatter(x[Ind], y[Ind])
            self.Iter = Ind
        
```

```python
Test = MinZero()
```

```python
Test.FindMin(30)
```
