```python
import numpy as np
import matplotlib.pyplot as plt
```

### Define the sigmoid activation function and its derivative
#### We define the sigmoid function, which maps input x to a value between 0 and 1, and its derivative, which represents the rate of change of the sigmoid function.

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))
```

### Generate input values and calculate the sigmoid and its derivative:
#### We create an array x of 1000 linearly spaced values between -5 and 5, and then compute the sigmoid and its derivative for each value in x.

```python
x = np.linspace(-5, 5, 1000)
sigmoid_x = sigmoid(x)
sigmoid_derivative_x = sigmoid_derivative(x)
```

### Multiply the sigmoid derivative 10 times:
#### We loop n times, multiplying the sigmoid derivative by itself at each step and storing the results in the gradient_values list. This simulates the backpropagation process in a deep neural network, where gradients are multiplied through layers.

```python
n= 5
gradient_values = []
for _ in range(n):
    sigmoid_derivative_x = sigmoid_derivative_x * sigmoid_derivative(x)
    gradient_values.append(sigmoid_derivative_x)
```

### Visualize the results:
#### We create a plot to visualize the gradient values at each step, clearly showing that as we multiply the sigmoid derivative more and more, the gradient values become extremely small, especially around the tails of the sigmoid function.

```python
plt.figure(figsize=(12, 6))
for i, value in enumerate(gradient_values):
    plt.plot(x, value, label=f'Step {i+1}')

plt.title('Vanishing Gradient Problem Demonstration')
plt.xlabel('x')
plt.ylabel('Gradient Value')
plt.legend()
plt.show()
```
